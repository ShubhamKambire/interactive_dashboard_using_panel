```python
#import Libraries
import pandas as pd
import numpy as np
import hvplot.pandas
import panel as pn
pn.extension('tabulator')
```

```python
video_games = pd.read_csv("vgsales.csv")
```

```python
#dropping na values
vg_cleaned = video_games.dropna(subset = ['Year'])
vg_cleaned.shape
```

```python
vg_cleaned.head(20)
```

```python
#creating dataframe for only sales and publisher values
publisher_values = vg_cleaned.groupby(['Year','Publisher'])[["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]].sum().reset_index()
publisher_values.head(30)
```

Creating Select Button for Publisher Values

```python
#Creating unique list of publisher
publisher_unique = list(publisher_values["Publisher"].unique())
publisher_unique
```

```python
#Make Dataframe interactive
publisher_int = publisher_values.interactive()
```

```python
#Select button for publisher values
publisher_select = pn.widgets.Select(name = "Publisher Names",options = publisher_unique)
publisher_select
```

```python
#Creating publisher pipeline
publisher_pipeline = (
    publisher_int[
        publisher_int["Publisher"] == publisher_select
    ]
    .sort_values(by='Year')  
    .reset_index(drop=True)
)
publisher_pipeline
```

Publisher Sales Over time

```python
#Line Graph for Sales of specific Publisher
Sales = ['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']
publisher_plot = publisher_pipeline.hvplot(x = 'Year',y = Sales,by = Sales,line_width = 2,title = "Publisher Sales")
publisher_plot
```

```python
#Creating another DataFrame for Sales according to Genre and Publisher
publisher_genre = vg_cleaned.groupby(['Publisher','Genre'])[["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]].sum().reset_index()
publisher_genre_int = publisher_genre.interactive()
```

Genre Based Sales over time

```python
publisher_genre_pipeline = (
    publisher_genre_int[publisher_int["Publisher"] == publisher_select].sort_values('Publisher').reset_index()
)
```

```python
publisher_genre_pipeline
```

```python
publisher_genre_plot = publisher_genre_pipeline.hvplot(kind = 'bar',x='Genre',y = 'Global_Sales',title = 'Genre-based sales trends over time')
publisher_genre_plot
```

Overall Sales Performance over time

```python
#Line Graph for overall Sales 
sales_data = vg_cleaned[["Year","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]].sort_values(by = "Year").reset_index()
sales_data = sales_data.groupby("Year")[Sales].sum().reset_index()
```

```python
sales_data_int = sales_data.interactive()
```

```python
sales_data_plot = sales_data_int.hvplot(x = "Year",y = Sales,line_width = 2,title = "Sales performance over time")
```

```python
sales_data_plot
```

```python
#Publisher Sales data in tabular format
publisher_table = publisher_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
publisher_table
```
