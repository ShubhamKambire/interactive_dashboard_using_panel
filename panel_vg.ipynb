```python
import pandas as pd
import numpy as np
import hvplot.pandas
import panel as pn
pn.extension('tabulator')
```

```python
video_games = pd.read_csv("vgsales.csv")
```

```python
vg_cleaned = video_games.dropna(subset = ['Year'])
vg_cleaned.shape
```

```python
vg_cleaned.head(20)
```

```python
publisher_values = vg_cleaned.groupby(['Year','Publisher'])[["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]].sum().reset_index()
publisher_values.head(30)
```

```python
publisher_unique = list(publisher_values["Publisher"].unique())
publisher_unique
```

```python
publisher_int = publisher_values.interactive()
```

```python
year_slider = pn.widgets.IntSlider(name = 'Year slider',start = 1980,end = 2020,step = 5,value = 2000)
year_slider
```

```python
publisher_select = pn.widgets.Select(name = "Publisher Names",options = publisher_unique)
publisher_select
```

```python

co2_pipeline = (
    publisher_int[
        publisher_int["Publisher"] == publisher_select
    ]
    .sort_values(by='Year')  
    .reset_index(drop=True)
)
co2_pipeline
```

```python
Sales = ['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']
publisher_plot = co2_pipeline.hvplot(x = 'Year',y = Sales,by = Sales,line_width = 2,title = "Publisher Sales")
publisher_plot
```

```python
genre_val = vg_cleaned.groupby(["Year","Genre"])[["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]].sum().reset_index()
```

```python
genre_int = genre_val.interactive()
```

```python

genre_select = pn.widgets.Select(name = "Genre's",options = list(genre_val["Genre"].unique()))
```

```python
genre_select
```

```python
genre_pipline = (
    genre_int[(genre_int.Year <= year_slider) & (genre_int["Genre"] == genre_select)] 
).sort_values("Year").reset_index()
```

```python
genre_plot = genre_pipline.hvplot(x = 'Year',y = Sales,by = Sales,line_width = 2,title = "Genre Sales")
genre_plot
```

```python
publisher_genre = vg_cleaned.groupby(['Publisher','Genre'])[["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]].sum().reset_index()
publisher_genre_int = publisher_genre.interactive()
```

```python

publisher_genre_select = pn.widgets.Select(name = "Publisher Names",options = list(publisher_genre["Publisher"].unique()))
```

```python
publisher_genre_select
```

```python
publisher_genre_pipeline = (
    publisher_genre_int[publisher_int["Publisher"] == publisher_select].sort_values('Publisher').reset_index()
)
```

```python
publisher_genre_pipeline
```

```python

publisher_genre_plot = publisher_genre_pipeline.hvplot(kind = 'bar',x='Genre',y = 'Global_Sales',title = 'Sales using Genre')
publisher_genre_plot
```

```python
vg_cleaned.head()
```

```python
sales_data = vg_cleaned[["Year","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]].sort_values(by = "Year").reset_index()
sales_data = sales_data.groupby("Year")[Sales].sum().reset_index()
```

```python
sales_data.head(20)
```

```python
sales_data_plot = sales_data.hvplot(x = "Year",y = Sales,line_width = 2,title = "Sales performance over time")
```

```python
sales_data_plot
```

```python
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
co2_table
```
